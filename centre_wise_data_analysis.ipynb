```python

for city_index in range(len(cities)):
    dfs=[]
    for yea in range(2002,2024):
        for month in range(1,13):
#months = range(1, 13)  # Assuming you want to access files from January (1) to December (12)
# Create an empty list to store the DataFrames

# Iterate over the months and load the CSV files

            file_name = os.path.join(str(yea), f'{month:02d}.csv')
            try:
                with open(file_name, 'r') as file:
                    pass

            except FileNotFoundError:
                continue
            file_name = file_name.replace('\\', '/')
            print(file_name)
            with open(file_name,'rb') as x:
                res=charset_normalizer.detect(x.read(1000000))
            df = pd.read_csv(file_name,encoding=res['encoding'],header=6)
            year = file_name.split('/')[0]  # Split the file name at '/' and take the first part as the year
            month = file_name.split('/')[1].split('.')[0]  # Split the file name at '/' and take the second part before '.' as the month
            y = int(year)
            m = int(month)
            day_columns = df.columns[df.columns.str.startswith('Day')]
            converted_columns = pd.to_datetime(day_columns.str.extract('(\d+)', expand=False), format='%d')
            modified_columns = (converted_columns + pd.offsets.DateOffset(month=m, year=y)).date
            column_mapping = {old_column: new_column for old_column, new_column in zip(day_columns, modified_columns)}
            df = df.rename(columns=column_mapping)
            df.Center=df.Center.fillna(method='ffill')
            df=df[df['Center'].str.startswith(cities[city_index])]
            if df.empty:
                pass
            else:
                df = df.dropna(subset=['Parameter'])
                df.loc[df['Parameter'].str.startswith('Arr')]
                df = df.dropna(subset=['Parameter'])
            #date_columns = df.columns[df.columns.get_loc('2002-04-01'):].tolist()
                s1 = df.loc[df['Parameter'].str.startswith('ArrivalQty')].squeeze()
                s2 = df.loc[df['Parameter'].str.startswith('Retail')].squeeze()
                s3=df.loc[df['Parameter'].str.startswith('Min')].squeeze()
                s4=df.loc[df['Parameter'].str.startswith('Max')].squeeze()
                s5=df.loc[df['Parameter'].str.startswith('Mod')].squeeze()
                data = {'Arrival Quantity': s1, 'Retail Prices': s2,'Min Price':s3,'Max Price':s4,'Modal Price':s5}
                df2 = pd.DataFrame(data)
                path_citiy=str(cities[city_index])+"/"+"latest_data.csv"
                folder=str(cities[city_index])
                print(path_citiy)
                os.makedirs(folder, exist_ok=True)
                df2.to_csv(path_citiy)
                df3=pd.read_csv(path_citiy,header=4,parse_dates=[0],index_col=0,names=['Date','Arrival Quantity','Retail Prices','Min Price','Max Price','Modal Price'])
                df3.isnull().sum()
                df3.info()
                #df3.to_csv('lucknowdata1.csv')
            if df.empty:
                print("DataFrame is empty")
            else:
                dfs.append(df3)
        if len(dfs)!=0:
            df45=pd.concat(dfs)
            path_2_city=cities[city_index]+"/"+"complete_data.csv"
            os.makedirs(str(cities[city_index]), exist_ok=True)
            df45.to_csv(path_2_city)
            print(df45)
        else:
            print("df45 is empty")
#df31=df3
```
